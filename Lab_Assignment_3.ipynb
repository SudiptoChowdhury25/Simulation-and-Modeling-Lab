## • Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]

## • Hypothesized mean: 5.5

## o Calculate the t-statistic and p-value.

## o Interpret and visualize the result at a 0.05 significance level.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

# Print t-statistic and p-value
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Interpretation at 0.05 significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from 5.5.")
else:
    print("Fail to reject the null hypothesis: No significant difference from 5.5.")

# Visualization
plt.figure(figsize=(8,5))
plt.hist(data, bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label=f"Sample Mean: {np.mean(data):.2f}")
plt.axvline(hypothesized_mean, color='green', linestyle='dashed', linewidth=2, label="Hypothesized Mean: 5.5")
plt.legend()
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data with Mean Comparison")
plt.show()
```
